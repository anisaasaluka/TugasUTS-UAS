```python
# ==============================
# CEK KELAYAKAN PRODUKSI
# WearEase Clothing
# ==============================

# Parameter kapasitas
kain_max = 300
tenaga_kerja_max = 160
mesin_max = 100

# Fungsi cek feasibility
def cek_produksi(x1, x2):
    kain = (3 * x1) + (1.5 * x2)
    tenaga_kerja = (2 * x1) + (1 * x2)
    mesin = (1.5 * x1) + (0.5 * x2)
    laba = (120000 * x1) + (30000 * x2)

    print(f"\n=== UJI PRODUKSI (Hoodie={x1}, T-shirt={x2}) ===")
    print(f"Kain         : {kain} / {kain_max}")
    print(f"Tenaga Kerja : {tenaga_kerja} / {tenaga_kerja_max}")
    print(f"Mesin        : {mesin} / {mesin_max}")
    print(f"Laba         : Rp {laba:,}")

    feasible = True

    if kain > kain_max:
        print(" Kendala kain DILANGGAR")
        feasible = False
    else:
        print(f" Kain terpenuhi (slack = {kain_max - kain})")

    if tenaga_kerja > tenaga_kerja_max:
        print(" Kendala tenaga kerja DILANGGAR")
        feasible = False
    else:
        print(f" Tenaga kerja terpenuhi (slack = {tenaga_kerja_max - tenaga_kerja})")

    if mesin > mesin_max:
        print(" Kendala mesin DILANGGAR")
        feasible = False
    else:
        print(f" Mesin terpenuhi (slack = {mesin_max - mesin})")

    if feasible:
        print(" KESIMPULAN: FEASIBLE (BALANCE)")
    else:
        print(" KESIMPULAN: TIDAK FEASIBLE")

# ------------------------------
# UJI 1 (Hasil Excel Solver)
# ------------------------------
cek_produksi(40, 80)

# ------------------------------
# UJI 2 (Tidak feasible)
# ------------------------------
cek_produksi(50, 90)
```
